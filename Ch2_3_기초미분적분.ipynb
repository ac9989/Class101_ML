#기초 미분적분학 (calculus)

기초미분

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

$y = x^3 +0.5x^2 - 0.7x + 0.1$

```python
x= np.linspace(-5,5,100)
def f1(x):
  return x**3+0.5*x**2-0.7*x+0.1

y = f1(x)
plt.plot(x,y) 
plt.grid() 
```

$ \frac{dy}{dx}\ = 3x^2 + x - 0.7$

```python
#미분(gradient) 자동 계산 method (수치적 방식)
dydx = np.gradient(y,x)
plt.plot(x,dydx)
plt.grid()
```

```python
# 해석적 미분(analytic, 인간이 푸는 방식)
dydx_ana = 3*x**2 + x - 0.7

plt.plot(x,dydx,'r*-',label='Num')
plt.plot(x,dydx_ana,'b-',label='ana')
plt.grid()
plt.legend()
```

- 다중 변수함수의 미분

$z = f(x_1 ,x_2)$

$x_1$과 $x_2$를 각각 미분하면? $ \frac{\partial z}{\partial x_1}\ ,\frac{\partial z}{\partial x_2}\ $

$ \frac{\partial z}{\partial x_1}\ = \lim_{h \to 0} \frac{f(x_1 + h,x_2)-f(x_1,x_2)}{h}\ $




$ f(x_1,x_2) = x_1x_2+3x_1^2-0.5x_2 $ 를  $x_1$,$x_2$로 미분

```python
x1 = np.linspace(-3,3)
x2 = np.linspace(-3,3)

def f(x1,x2):
  return x1*x2+3*x1**2-0.5*x2

def grad_x1(x1,x2):
  return (f(x1+1e-10,x2)-f(x1,x2))/1e-10  # h=1e-10 = 10의 -10승

def grad_x2(x1,x2):
  return (f(x1,x2+1e-10)-f(x1,x2))/1e-10

```

```python
grad1=grad_x1(x1,3)
plt.plot(x1,grad1)
plt.grid()
```

```python
grad2=grad_x2(0.5,x2)
plt.plot(x2,grad2)
plt.grid()
```

```python
#과제
x1 = 2 
x2 = np.linspace(-3,3)

def grad2(x1,x2):
  return 5*x1*x2+x1**2+2.3

grad=grad2(2,x2)
plt.plot(x2,grad)
plt.grid()
```
