- 텐서
 1. 여기부터 어렵대여
 2. 다중선형사상
 3. 벡터가 쌓이는 것 = 랭크의 추가


_________________
- 텐서의 기초연산
 1. 덧셈
 2. 곱셈
   - 행렬 곱셈
   - 원소별 곱셈

```python
#덧셈 => 같은 형태의 텐서끼리만 덧셈 가능
A = np.array([3,5,6])
B = np.array([-1,2,0])
print(A+B)
```

```python
print(A-B)
B *= -1
print(A+B)
```

```python
A_1 = np.array([3,5,6])
A_2 = np.array([3,5,6,0])
print(A_1+A_2)   # => 차원이 다른 두 텐서(벡터)는 연산할 수 없음
```

```python
# 행렬 곱셈 => 앞텐서의 마지막 차원, 뒷텐서의 첫번째차원이 같아야 계산 가능
A = np.random.normal(size=(5,3))
B = np.random.normal(size=(3,6))
print(A.shape)
print(B.shape)
C = np.matmul(A,B)
print(C.shape)
```

```python
#과제
A_1 = np.random.normal(size=(10,5,3))
B_1 = np.random.normal(size=(3,20))
C_1 = np.matmul(A_1,B_1)
print(C_1.shape)
print(C_1)
```

```python
# 원소의 곱셈 => 두 텐서의 형태가 동일할 때 가능한 곱셈
A = np.array([[1,3,5],[2,4,8]])
B = np.array([[-1,0,3],[5,6,-2]])
C = np.multiply(A,B)
print('값',C,", 형태",C.shape)
```

- 텐서 조작
 1. 재배열 (스칼라=>벡터, 벡터=>행렬 등등 랭크를 높이는 행위)
 2. 텐서 합치기

```python
#재배열 numpy => reshape , torch => view
a= np.array(30)
print("형태",a.shape) #()
a_new = a.reshape(-1,1) # -1*1 행렬, 숫자가 하나 뿐이므로 -1은 1로 치환됨
print("형태",a_new.shape)

vec1 = np.linspace(-1,1,100) #100의 벡터
print('형태',vec1.shape)
#(100,) => (5,20)
mat1_vec_1 = vec1.reshape(-1,20)  # 5 X 20 = 100이므로 -1이 5로 치환됨
print('바뀐형태', mat1_vec_1.shape)

# 행렬 재배열 Transpose 즉 전치행렬
a = np.random.normal(size=(3,5))
b = a.T 
print('a 형태',a.shape)
print('a.T 형태',b.shape)
```

```python
vec1 = np.array([1,2,3])
vec2 = np.array([2,4,5])
vec3 = np.array([3,6,10])

# 백터의 연결 
#1번째 3+3+3 (9,)
con1 = np.concatenate((vec1,vec2,vec3))
print('9차원 벡터 결과',con1.shape)
#2번째 (3,1),(3,1),(3,1) (3,3)
con2 = np.concatenate((vec1.reshape(-1,1),vec2.reshape(-1,1),vec3.reshape(-1,1)),axis=1) #더해야할 차원의 옵션 즉 (3,1)에서 1을 더하기 위해 두번째(=1) 차원을 더함
print('3*3차원 벡터 결과',con2.shape)
```

```python
#과제2
vec1 = np.array([1,2,3]).reshape(-1,1)
vec2 = np.array([2,4,5]).reshape(-1,1)
vec3 = np.array([3,6,10]).reshape(-1,1)
con3 = np.concatenate((vec1,vec2,vec3),axis=1)
print(con3.shape)
```
